```r
#all
#Read the PRS scores file
prs_scores <- read.table("pretrained_pgs_all.profile", header = TRUE)
```

```r


#Read the height phenotype file
height_data <- read.table("height_formatted.txt", header = TRUE)
```

```r
merged_all <- merge(prs_scores, height_data, by = "IID")
```

```r
merged_all
```

```r
# Predict height using a linear regression model
model <- lm(Height ~ SCORESUM, data = merged_all)
summary(model)
```

```r
# predict
merged_all$predicted_height <- predict(model, newdata = merged_all)
merged_all
```

```r
library(ggplot2)

ggplot(merged_all, aes(x = predicted_height, y = Height)) +
  geom_point(color = "blue", alpha = 0.5) +
  geom_abline(slope = 1, intercept = 0, color = "red", linetype = "dashed") +
  labs(title = "Actual Height vs Predicted Height",
       x = "Predicted Height",
       y = "Actual Height") +
  theme_minimal()
```

```r

errors <- merged_all$Height - merged_all$predicted_height

#mse
mse <- mean(errors^2)
cat("Mean Squared Error (MSE):", mse, "\n")

# RMSE
rmse <- sqrt(mse)
cat("Root Mean Squared Error (RMSE):", rmse, "\n")

# MAE
mae <- mean(abs(errors))
cat("Mean Absolute Error (MAE):", mae, "\n")

# R²
r_squared <- summary(model)$r.squared
cat("R² value:", r_squared, "\n")
```
