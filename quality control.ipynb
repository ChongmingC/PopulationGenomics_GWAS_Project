```r
het = read.table('gwas_data.het', header = T)
```

```r
het
```

```r
het$het_cal = (het$N.NM. - het$O.HOM.)/ het$N.NM.
```

```r
OHOM_zero_rows <- het[het$O.HOM == 0, ]
# some O.HOM. are 0, we need exclude these samples from analysis
```

```r
OHOM_zero_rows
```

```r
het1 <- het[het$O.HOM !=0,]
```

```r
het1
```

```r
mean_het = mean(het1$het_cal)
sd_het = sd(het1$het_cal)
```

```r
mean_het
```

```r
sd_het
```

```r
#0.03
plot(x = het$het_cal,y = Imiss$F_MISS, lwd =4, col = 'light blue', xlab = 'Heterozygosity rate per individual', ylab = 'Proportion of missing SNPs per individual')
abline(h = 0.03, col = 'red', lty = 3)
abline(v = mean_het + 3*sd_het, col = 'red', lty = 3)
abline(v = mean_het - 3*sd_het, col = 'red', lty = 3)
```

```r
#0.09
plot(x = het$het_cal,y = Imiss$F_MISS, lwd =4, col = 'light blue', xlab = 'Heterozygosity rate per individual', ylab = 'Proportion of missing SNPs per individual')
abline(h = 0.9, col = 'red', lty = 3)
abline(v = mean_het + 3*sd_het, col = 'red', lty = 3)
abline(v = mean_het - 3*sd_het, col = 'red', lty = 3)
```

```r
#proceed to filter samples based on these criteria
```

```r
right_tail = mean_het + 3*sd(het1$het_cal)
left_tail =  mean_het - 3*sd(het1$het_cal)
```

```r
filtering = cbind(Imiss, het)
outlier_ind = subset(filtering, filtering$F_MISS >= 0.9 | filtering$het_cal > right_tail | filtering$het_cal < left_tail)

```

```r
outlier_ind
```

```r
nrow(outlier_ind)
```

```r
write.table(outlier_ind[,c(1,2)], 'wrong_het_missing_values.txt', col.names = FALSE, row.names = FALSE)
```

Genomic Control.T The inflation factor (𝝺) can be calculated as the median of the Chi-squared statistics computed divided by the median of the Chi-squared distribution under the null. Given a p-value (p) the corresponding Chi-squared quantile can be calculated as:

```r
p <- 0.5
```

```r
## 计算理论上的χ²统计量中位数（在无关联假设下为0.456）
theoretical_median <- qchisq(0.5, df = 1)
```

```r
# 计算χ²统计量的中位数
median_chi_squared <- median(qchisq(1-d$P, df=1))
```

```r
lambda <- median_chi_squared / theoretical_median
```
